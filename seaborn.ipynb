#seaborn

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
var1 = [1,2,3,4,5,6,7,8]
var2 = [8,7,6,5,3,9,9,8]
```

```python
#matplot
plt.plot(var1, var2)
plt.show()
```

```python
#seaborn

sns.lineplot(x = var1 ,y = var2)
plt.show()
```

```python
var1 = [1,2,3,4,5,6,7,8]
var2 = [8,7,6,5,3,9,9,8]

import pandas as pd

df = pd.DataFrame({'var1':var1, 'var2':var2})
sns.lineplot(x = 'var1', y = 'var2', data = df)
plt.show()
```

```python
df1 = sns.load_dataset("penguins")
```

```python
df1
```

```python
sns.lineplot(x = 'bill_length_mm',  y = 'flipper_length_mm', data = df1)
plt.show()
```

```python
sns.lineplot(x = 'bill_length_mm',  y = 'flipper_length_mm', data = df1, hue = 'sex')
plt.show()
```

```python
sns.lineplot(x = 'bill_length_mm',  y = 'flipper_length_mm', data = df1, hue = 'sex', palette = "rocket_r")
plt.show()
```

```python
sns.lineplot(x = 'bill_length_mm',  y = 'flipper_length_mm', data = df1, hue = 'sex',
             palette = "rocket_r", markers = ["o",">"])
plt.show()
```
